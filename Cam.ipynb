```python
####WORKING
# from tensorflow.keras.preprocessing.image import img_to_array
# from tensorflow.keras.models import load_model
# import numpy as np
# import cv2
# import cvlib as cv

# # Load model
# try:
#     model = load_model('gender_detection.h5')  # Adjust if using a different format
# except Exception as e:
#     print(f"Error loading model: {e}")
#     exit()

# # Open webcam
# webcam = cv2.VideoCapture(0)

# # Check if webcam opened successfully
# if not webcam.isOpened():
#     print("Error: Could not open webcam.")
#     exit()

# classes = ['man', 'woman']

# # Set to keep track of counted faces
# counted_faces = set()

# # Loop through frames
# while webcam.isOpened():
#     # Read frame from webcam 
#     status, frame = webcam.read()
#     if not status:
#         print("Error: Could not read frame.")
#         break

#     # Apply face detection
#     faces, confidence = cv.detect_face(frame)

#     # Initialize counters for men and women for this frame
#     count_men = 0
#     count_women = 0

#     # Reset the set of counted faces for this frame
#     current_frame_faces = set()

#     # Loop through detected faces
#     for idx, f in enumerate(faces):
#         # Get corner points of face rectangle        
#         (startX, startY) = f[0], f[1]
#         (endX, endY) = f[2], f[3]

#         # Create a unique identifier for the face based on its bounding box
#         face_id = (startX, startY, endX, endY)

#         # Check if this face has already been counted in previous frames
#         if face_id in counted_faces:
#             continue  # Skip counting this face again

#         # Draw rectangle over face
#         cv2.rectangle(frame, (startX, startY), (endX, endY), (0, 255, 0), 2)

#         # Crop the detected face region
#         face_crop = np.copy(frame[startY:endY, startX:endX])

#         if face_crop.shape[0] < 10 or face_crop.shape[1] < 10:
#             continue

#         # Preprocessing for gender detection model
#         face_crop = cv2.resize(face_crop, (96, 96))
#         face_crop = face_crop.astype("float") / 255.0
#         face_crop = img_to_array(face_crop)
#         face_crop = np.expand_dims(face_crop, axis=0)

#         # Apply gender detection on face
#         conf = model.predict(face_crop)[0]  # Ensure this matches your model output

#         # Get label with max accuracy
#         idx = np.argmax(conf)
#         label = classes[idx]

#         # Count men and women for this face
#         if label == 'man':
#             count_men += 1
#         else:
#             count_women += 1

#         # Mark this face as counted
#         counted_faces.add(face_id)

#         label = "{}: {:.2f}%".format(label, conf[idx] * 100)

#         Y = startY - 10 if startY - 10 > 10 else startY + 10

#         # Write label and confidence above face rectangle
#         cv2.putText(frame, label, (startX, Y), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

#         # Add the current face ID to the current frame's set
#         current_frame_faces.add(face_id)

#     # Update the counted faces set with the current frame's faces
#     counted_faces.update(current_frame_faces)

#     # Display the counts on the frame for this frame only
#     cv2.putText(frame, f'Men: {count_men}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
#     cv2.putText(frame, f'Women: {count_women}', (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

#     # Display output
#     cv2.imshow("Gender Detection", frame)

#     # Press "Q" to stop
#     if cv2.waitKey(1) & 0xFF == ord('q'):
#         break

# # Release resources
# webcam.release()
# cv2.destroyAllWindows()

#//////////////////////////////////////////////////////////////////////////



## Working code With ratio
# from tensorflow.keras.preprocessing.image import img_to_array
# from tensorflow.keras.models import load_model
# import numpy as np
# import cv2
# import cvlib as cv

# # Load model
# try:
#     model = load_model('gender_detection.h5')  # Adjust if using a different format
# except Exception as e:
#     print(f"Error loading model: {e}")
#     exit()

# # Open webcam
# webcam = cv2.VideoCapture(0)

# # Check if webcam opened successfully
# if not webcam.isOpened():
#     print("Error: Could not open webcam.")
#     exit()

# classes = ['man', 'woman']

# # Loop through frames
# while webcam.isOpened():
#     # Read frame from webcam 
#     status, frame = webcam.read()
#     if not status:
#         print("Error: Could not read frame.")
#         break

#     # Apply face detection
#     faces, confidence = cv.detect_face(frame)

#     # Initialize counters for men and women for this frame
#     total_men = 0
#     total_women = 0

#     # Loop through detected faces
#     for idx, f in enumerate(faces):
#         # Get corner points of face rectangle        
#         (startX, startY) = f[0], f[1]
#         (endX, endY) = f[2], f[3]

#         # Draw rectangle over face
#         cv2.rectangle(frame, (startX, startY), (endX, endY), (0, 255, 0), 2)

#         # Crop the detected face region
#         face_crop = np.copy(frame[startY:endY, startX:endX])

#         if face_crop.shape[0] < 10 or face_crop.shape[1] < 10:
#             continue

#         # Preprocessing for gender detection model
#         face_crop = cv2.resize(face_crop, (96, 96))
#         face_crop = face_crop.astype("float") / 255.0
#         face_crop = img_to_array(face_crop)
#         face_crop = np.expand_dims(face_crop, axis=0)

#         # Apply gender detection on face
#         conf = model.predict(face_crop)[0]  # Ensure this matches your model output

#         # Get label with max accuracy
#         idx = np.argmax(conf)
#         label = classes[idx]

#         # Count men and women for this face
#         if label == 'man':
#             total_men += 1
#         else:
#             total_women += 1

#         label = "{}: {:.2f}%".format(label, conf[idx] * 100)

#         Y = startY - 10 if startY - 10 > 10 else startY + 10

#         # Write label and confidence above face rectangle
#         cv2.putText(frame, label, (startX, Y), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

#     # Calculate real-time ratio of men to women
#     if total_women > 0:
#         ratio = total_men / total_women
#         ratio_text = f'Ratio (Men:Women) = {ratio:.2f}:1'
#     else:
#         ratio_text = 'Ratio (Men:Women) = N/A'

#     # Display the counts on the frame for this frame only
#     cv2.putText(frame, f'Men: {total_men}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
#     cv2.putText(frame, f'Women: {total_women}', (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

#     # Display the real-time ratio
#     cv2.putText(frame, ratio_text, (10, 90), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

#     # Display output
#     cv2.imshow("Gender Detection", frame)

#     # Press "Q" to stop
#     if cv2.waitKey(1) & 0xFF == ord('q'):
#         break

# # Release resources
# webcam.release()
# cv2.destroyAllWindows()


#//////////////////////////////////////////////////////////////////////////////////////////////


from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model
import numpy as np
import cv2
import cvlib as cv
from twilio.rest import Client
import ctypes
import requests

# Function to get screen resolution
def get_screen_resolution():
    user32 = ctypes.windll.user32
    screen_width = user32.GetSystemMetrics(0)
    screen_height = user32.GetSystemMetrics(1)
    return screen_width, screen_height

# Twilio configuration
account_sid = 'AC1e7af39842335d02166094adac006d45'  # Replace with your Twilio account SID
auth_token = 'c3cf200f6d24029926c04b69a79669db'    # Replace with your Twilio Auth Token
twilio_number = '+12088455467'  # Replace with your Twilio phone number

# List of phone numbers to send SMS to
phone_numbers = ['+917447495973', '+917385268226']  # Replace with the phone numbers to receive alerts

client = Client(account_sid, auth_token)

# Function to get accurate location coordinates from IP
def get_location():
    try:
        # Fetch location data based on the public IP address
        response = requests.get('https://ipinfo.io/json')
        data = response.json()

        # Extract location details
        location = data.get('loc')  # Latitude and longitude in the form "lat,long"
        city = data.get('city')
        region = data.get('region')
        country = data.get('country')

        # Format location details
        if location:
            lat, lon = location.split(',')
            location_text = f'{city}, {region}, {country} (Lat: {lat}, Long: {lon})'
        else:
            location_text = f'{city}, {region}, {country} (Lat, Long: Not available)'

        return location_text
    except Exception as e:
        print(f"Error retrieving location: {e}")
        return "Location unavailable"

# Function to send SMS alert with location to multiple numbers
def send_sms_alert_with_location(ratio):
    location = get_location()
    message_body = f'HELP!!!!!!!!!\nLocation: {location}'
    for number in phone_numbers:
        try:
            message = client.messages.create(
                body=message_body,
                from_=twilio_number,
                to=number
            )
            print(f"SMS alert sent to {number}: {message.sid}")
        except Exception as e:
            print(f"Error sending SMS to {number}: {e}")

# Load model
try:
    model = load_model('gender_detection.h5')
except Exception as e:
    print(f"Error loading model: {e}")
    exit()

# Open webcam
webcam = cv2.VideoCapture(0)

# Check if webcam opened successfully
if not webcam.isOpened():
    print("Error: Could not open webcam.")
    exit()

classes = ['man', 'woman']
threshold_ratio = 2.0  # Ratio threshold for sending SMS alert

# Get screen resolution
screen_res = get_screen_resolution()

# Create a fullscreen window
cv2.namedWindow("Gender Detection", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Gender Detection", screen_res[0], screen_res[1])
cv2.setWindowProperty("Gender Detection", cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

# Loop through frames
while webcam.isOpened():
    # Read frame from webcam 
    status, frame = webcam.read()
    if not status:
        print("Error: Could not read frame.")
        break

    # Resize frame to match screen resolution
    frame = cv2.resize(frame, (screen_res[0], screen_res[1]))

    # Apply face detection
    faces, confidence = cv.detect_face(frame)

    # Initialize counters for men and women for this frame
    total_men = 0
    total_women = 0

    # Loop through detected faces
    for idx, f in enumerate(faces):
        # Get corner points of face rectangle        
        (startX, startY) = f[0], f[1]
        (endX, endY) = f[2], f[3]

        # Draw rectangle over face
        cv2.rectangle(frame, (startX, startY), (endX, endY), (0, 255, 0), 2)

        # Crop the detected face region
        face_crop = np.copy(frame[startY:endY, startX:endX])

        if face_crop.shape[0] < 10 or face_crop.shape[1] < 10:
            continue

        # Preprocessing for gender detection model
        face_crop = cv2.resize(face_crop, (96, 96))
        face_crop = face_crop.astype("float") / 255.0
        face_crop = img_to_array(face_crop)
        face_crop = np.expand_dims(face_crop, axis=0)

        # Apply gender detection on face
        conf = model.predict(face_crop)[0]  # Ensure this matches your model output

        # Get label with max accuracy
        idx = np.argmax(conf)
        label = classes[idx]

        # Count men and women for this face
        if label == 'man':
            total_men += 1
        else:
            total_women += 1

        label = "{}: {:.2f}%".format(label, conf[idx] * 100)

        Y = startY - 10 if startY - 10 > 10 else startY + 10

        # Write label and confidence above face rectangle
        cv2.putText(frame, label, (startX, Y), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    # Calculate real-time ratio of men to women
    if total_women > 0:
        ratio = total_men / total_women
        ratio_text = f'Ratio (Men:Women) = {ratio:.2f}:1'
        
        # Check if ratio exceeds threshold and send alert
        if ratio > threshold_ratio:
            send_sms_alert_with_location(ratio)
    else:
        ratio = float('inf')
        ratio_text = 'Ratio (Men:Women) = N/A'

    # Display the counts on the frame for this frame only
    cv2.putText(frame, f'Men: {total_men}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    cv2.putText(frame, f'Women: {total_women}', (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    # Display the real-time ratio
    cv2.putText(frame, ratio_text, (10, 90), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    # Display output in fullscreen
    cv2.imshow("Gender Detection", frame)

    # Press "Q" to stop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
webcam.release()
cv2.destroyAllWindows()














```
